### E. 
Plotting normal random numbers
and show that addition of two Gaussian is Gaussian 
and addition of all is Gaussian

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

x1 = np.random.normal(0, 1, 1000)
x2 = np.random.normal(0, 1, 1000)
x3 = np.random.normal(0, 1, 1000)
x4 = np.random.normal(0, 1, 1000)
x5 = np.random.normal(0, 1, 1000)
x6 = np.random.normal(0, 1, 1000)
x7 = np.random.normal(0, 1, 1000)

sum_two = x1 + x2
sum_all = x1 + x2 + x3 + x4 + x5 + x6 + x7

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(x1, bins=30, alpha=0.5, label='x1')
plt.hist(x2, bins=30, alpha=0.5, label='x2')
plt.hist(sum_2, bins=30, alpha=0.7, label='x1 + x2', color='orange')
plt.legend()
plt.title('Sum of Two Gaussian Distributions')

plt.subplot(1, 2, 2)
plt.hist(sum_all, bins=30, alpha=0.7, label='Sum of All Gaussian', color='green')
plt.legend()
plt.title('Sum of Many Gaussian Distributions')
plt.tight_layout()
plt.show()
```
